```python
# Answer 1
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

output_notebook()

# Create data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a Bokeh plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='y')
p.line(x, y, line_width=2, color='blue', legend_label='sin(x)')

show(p)
```

```python
# Answer 2

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np
import pandas as pd

output_notebook()

# Create random data
np.random.seed(42)
n = 100
x = np.random.uniform(0, 10, n)
y = np.random.uniform(0, 10, n)
sizes = np.random.uniform(5, 50, n)  # Marker sizes
colors = np.random.choice(['red', 'green', 'blue', 'orange'], n)  # Marker colors

# Create a Bokeh plot
p = figure(title="Scatter Plot with Random Data", x_axis_label='x', y_axis_label='y')
p.scatter(x, y, size=sizes, color=colors, alpha=0.6)

show(p)
```

```python
# Answer 3

from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the output to display in a Jupyter notebook
output_notebook()

# Data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a new figure
p = figure(x_range=fruits, plot_height=350, title="Fruit Counts",
           toolbar_location=None, tools="")

# Add a bar renderer
p.vbar(x=fruits, top=counts, width=0.9)

# Customize the plot
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.axis.minor_tick_line_color = None
p.outline_line_color = None
p.xaxis.axis_label = "Fruits"
p.yaxis.axis_label = "Counts"
p.xaxis.major_label_orientation = 1

# Show the plot
show(p)
```

```python
# Answer 4

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the output to display in a Jupyter notebook
output_notebook()

# Generate random data
data_hist = np.random.randn(1000)

# Compute histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a new figure
p = figure(title="Histogram of Random Data", tools="", background_fill_color="#fafafa")

# Add a quad renderer for the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.7)

# Customize the plot
p.xaxis.axis_label = "Value"
p.yaxis.axis_label = "Frequency"
p.grid.grid_line_color = "white"

# Show the plot
show(p)
```
